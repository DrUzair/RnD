```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('C:/RnD/TSX/code/market_summary.csv')
list(df)
```

```python
cap_cut = 1e8
vol_cut = 1e5
df_cap_1e8_vol_1e6 = df[(df.market_cap > cap_cut) & (df.quote_volume > vol_cut)]
```

```python
top_loosers = df_cap_1e8_vol_1e6[df_cap_1e8_vol_1e6['price_precent_change_5d'] < -10]
top_loosers_ = top_loosers[['quote_symbol', 'price_precent_change_5d']]
top_loosers_.set_index("quote_symbol",drop=True,inplace=True)
top_loosers_.sort_values(by=['price_precent_change_5d'], ascending=True).plot.bar(figsize=(15,6))
plt.title('price_precent_change_5d -VE')
plt.show()
```

```python
top_gainers = df_cap_1e8_vol_1e6[df_cap_1e8_vol_1e6['price_precent_change_5d'] > 10]
top_gainers_ = top_gainers[['quote_symbol', 'price_precent_change_5d']]
top_gainers_.set_index("quote_symbol",drop=True,inplace=True)
top_gainers_.sort_values(by=['price_precent_change_5d'], ascending=True).plot.bar(figsize=(15,6))
plt.title('price_precent_change_5d +VE')
plt.show()
```
